```python
import json
import pandas as pd
from collections import OrderedDict
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

### 1. labels

```python
labels_file = open('../../../cache/labels.json', 'r', encoding='utf-8')
label_dicts = OrderedDict()
for line in labels_file:
    label = json.loads(line)
    label_dicts[label['label']] = label['label_des']
label_dicts
```

### 2. train dataset

```python
def json_to_df(json_file):
    data_file = open(json_file, 'r', encoding='utf-8')
    instances = []
    for line in data_file:
        instance = json.loads(line)
        instances.append([instance['label'], instance['label_des'], instance['sentence']])
    print(f'{json_file} dataset size: {len(instances)}')
    df = pd.DataFrame(instances, columns=['label', 'label_des', 'sentence'])
    return df
```

```python
train_df = json_to_df('../../../cache/train.json')
train_df
```

```python
print('train label frequency')
train_df_vcs = train_df['label_des'].value_counts()
for name, value in zip(train_df_vcs.index, train_df_vcs):
    print(f'{name}\t: {value}')
```

### 4. cross valid fix labels

```python
file0 = open('../../../cache/new_epoch6/test_prediction_fold_0.json', 'r')
file1 = open('../../../cache/new_epoch6/test_prediction_fold_1.json', 'r')
file2 = open('../../../cache/new_epoch6/test_prediction_fold_2.json', 'r')
file3 = open('../../../cache/new_epoch6/test_prediction_fold_3.json', 'r')
file4 = open('../../../cache/new_epoch6/test_prediction_fold_4.json', 'r')
file = open('../../../cache/new_epoch6/train_dev.json', 'r')

lables0 = [json.loads(line)['label'] for line in file0]
lables1 = [json.loads(line)['label'] for line in file1]
lables2 = [json.loads(line)['label'] for line in file2]
lables3 = [json.loads(line)['label'] for line in file3]
lables4 = [json.loads(line)['label'] for line in file4]
labels, labels_des, sentences = [], [], []
for line in file:
    labels.append(json.loads(line)['label'])
    labels_des.append(json.loads(line)['label_des'] )
    sentences.append(json.loads(line)['sentence'])
```

```python
results_error = []
results = []
for label0, label1, label2, label3, label4, label, label_des, sentence in zip(lables0, lables1, lables2, lables3, lables4, labels, labels_des, sentences):
    if len(set([label0, label1, label2, label3, label4])) == 1:  # 5fold预测全一致
        if label != label0:                                      # 标签与预测不一致
            results_error.append((label_dicts[label0], label0, label1, label2, label3, label4, label, label_des, sentence))
            label = label0                                       # 修改标签
    results.append({"label": label, "label_des": label_des, "sentence": sentence})
```

```python
with open('./train_dev_fixed.json', 'w', encoding='utf-8') as fp:
    for item in results:
        fp.write(json.dumps(item, ensure_ascii=False))
        fp.write('\n')
```
