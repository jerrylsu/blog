### 特殊单字符

- **.** :


- **\d** : [0-9]

- **\w** : [0-9a-zA-Z_]

- **\s** : [\r\n\t\f\v ]


- **\D** : [^0-9]

- **\W** : [^0-9a-zA-Z_]

- **\S** : [^\r\n\t\f\v ]


### 空白符

**\r \n \f \t \v 空格**  等价 **\s**

### 量词

**\*** : 出现0到多次

**+** : 出现1到多次

**?** : 出现0到1次

**{m}** :  出现m次

**{m,}** : 出现至少m次

**{m,n}** : 出现m到n次


### 范围

**|** : 或

**[ ]** : 多选一

**[^ ]** : 取反，不包含[ ]中任意字符
